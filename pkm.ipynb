```python
from sklearn.pipeline import Pipeline, make_pipeline, make_union
```

```python
from skrub import SelectCols
```

```python
num_features = ['Total', 'HP', 'Attack', 'Defense', 'Spatk', 'Spdef', 'Speed']
features = num_features + ['Type1', 'Type2']
```

```python
type2_pipeline = make_pipeline(SelectCols('Type2'), SimpleImputer(missing_values='nan', strategy='constant', fill_value='Pure'), OneHotEncoder(handle_unknown='infrequent_if_exist', sparse_output=False))
type1_pipeline = make_pipeline(SelectCols('Type1'), OneHotEncoder(handle_unknown='infrequent_if_exist',sparse_output=False))
numeric_pipeline = make_pipeline(SelectCols(num_features), StandardScaler())
```

```python
type2_pipeline
```

```python
type1_pipeline
```

```python
Features = make_union(type2_pipeline, type1_pipeline, numeric_pipeline)
```

```python
Features
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[features], df['Legendary'], test_size=0.1, random_state=42)
le_encoder = LabelEncoder()
y_train_e, y_test_e = le_encoder.fit_transform(y_train), le_encoder.transform(y_test)
```

```python
# Features.fit_transform(X_train, y_train_e)
```

```python
e2e_pipe = make_pipeline(Features, HistGradientBoostingClassifier())
```

```python
e2e_pipe
```

```python
y_pred = e2e_pipe.fit(X_train, y_train_e).predict(X_test)
```

```python
y_pred
```

```python
print(classification_report(y_test_e, y_pred))
```
